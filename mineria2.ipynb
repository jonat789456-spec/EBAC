```python
import pandas as pd
import re

def limpiar_caracteres_especiales(df: pd.DataFrame) -> pd.DataFrame:
    """
    Recorre todo el DataFrame, incluidos los encabezados,
    y elimina caracteres especiales reemplazándolos por espacios en blanco.
    """
    # Limpiar los nombres de columnas
    df.columns = [re.sub(r'[^A-Za-z0-9]+', ' ', str(col)).strip() for col in df.columns]
    
    # Limpiar cada celda del DataFrame
    df_limpio = df.copy()
    for col in df_limpio.columns:
        df_limpio[col] = df_limpio[col].apply(
            lambda x: re.sub(r'[^A-Za-z0-9]+', ' ', str(x)).strip() if pd.notnull(x) else x
        )
    
    return df_limpio



df = pd.DataFrame(df_muestra)
df_limpio_muestra = limpiar_caracteres_especiales(df)
```

```python
import arff  # pip install liac-arff

def guardar_arff(df, nombre_archivo):
    """
    Convierte un DataFrame de Pandas a un archivo .arff

    Parámetros:
    df : pd.DataFrame -> el DataFrame a convertir
    nombre_archivo : str -> nombre del archivo de salida (sin extensión)
    """
    # Definir los atributos (nombre y tipo)
    atributos = []
    for col in df.columns:
        if df[col].dtype == "object":
            # Para columnas categóricas, se listan los valores únicos
            valores_unicos = df[col].dropna().unique().tolist()
            atributos.append((col, valores_unicos))
        else:
            # Para columnas numéricas, se usa tipo real
            atributos.append((col, "REAL"))

    # Convertir filas a listas
    data = df.values.tolist()

    # Crear diccionario ARFF
    arff_dict = {
        "description": f"Archivo generado a partir del DataFrame {nombre_archivo}",
        "relation": nombre_archivo,
        "attributes": atributos,
        "data": data,
    }

    # Guardar archivo
    with open(f"{nombre_archivo}.arff", "w", encoding="utf-8") as f:
        arff.dump(arff_dict, f)
```

# ID3


```python
import pandas as pd
from sklearn.compose import ColumnTransformer, make_column_selector as selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ===== 1) Configuración =====
TARGET = "ProductivityLoss"  # ajusta si tu columna objetivo tiene otro nombre

df_train = df_limpio_entrenamiento_sin.copy()
df_test  = df_limpio_muestra_sin.copy()

# Verificación básica (opcional)
assert TARGET in df_train.columns and TARGET in df_test.columns, "Falta la columna objetivo en train/test."

X_train = df_train.drop(columns=[TARGET])
y_train = df_train[TARGET]

X_test  = df_test.drop(columns=[TARGET])
y_test  = df_test[TARGET]

# ===== 2) Preprocesamiento =====
preprocess = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"),
         selector(dtype_include=["object", "category"])),
        ("num", "passthrough",
         selector(dtype_exclude=["object", "category"]))
    ],
    remainder="drop",
    verbose_feature_names_out=False
)

# ===== 3) Modelo (ID3 ≈ entropía) =====
id3 = DecisionTreeClassifier(
    criterion="entropy",     # clave para emular ID3
    max_depth=None,          # ajusta p/evitar sobreajuste (ej. 6, 10, ...)
    random_state=42
    # class_weight="balanced" # Úsalo si hay desbalance severo de clases
)

clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", id3)
])

# ===== 4) Entrenar =====
clf.fit(X_train, y_train)

# ===== 5) Evaluar en TEST =====
y_pred = clf.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

print("Matriz de confusión (orden de labels detectados):")
labels = sorted(pd.unique(pd.concat([y_train, y_test], ignore_index=True)))
print(confusion_matrix(y_test, y_pred, labels=labels))
print("Labels:", labels)

# ===== 6) (Opcional) Árbol en texto =====
feature_names = clf.named_steps["prep"].get_feature_names_out()
tree_text = export_text(clf.named_steps["model"], feature_names=list(feature_names))
print("\nÁrbol (resumen en texto):\n")
print(tree_text)


# Accuracy en entrenamiento
train_acc = accuracy_score(y_train, clf.predict(X_train))
train_err = 1 - train_acc

# Accuracy en prueba
test_acc = accuracy_score(y_test, clf.predict(X_test))
test_err = 1 - test_acc

print(f"Entrenamiento -> Acc: {train_acc:.4f} | Error: {train_err:.4f}")
print(f"Prueba        -> Acc: {test_acc:.4f} | Error: {test_err:.4f}")
```

# ONE RULE 

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ===== 1) Configuración =====
TARGET = "ProductivityLoss"

df_train = df_limpio_entrenamiento_sin.copy()
df_test  = df_limpio_muestra_sin.copy()

X_train = df_train.drop(columns=[TARGET])
y_train = df_train[TARGET]

X_test  = df_test.drop(columns=[TARGET])
y_test  = df_test[TARGET]

# Clase mayoritaria global (para valores no vistos / NaN mapeados)
majority_class = y_train.value_counts().idxmax()

# ===== 2) Encontrar la mejor regla (atributo con menos errores) =====
best_feature = None
best_mapping = None
best_acc = -1.0

for col in X_train.columns:
    # Mapa valor -> clase mayoritaria (en entrenamiento) CORREGIDO: usar df_train
    mapping = (
        df_train.groupby(col)[TARGET]
        .agg(lambda s: s.value_counts().idxmax())
        .to_dict()
    )

    # Predicción en entrenamiento usando la regla de 1 columna
    y_pred_train = X_train[col].map(mapping).fillna(majority_class)

    acc = accuracy_score(y_train, y_pred_train)
    if acc > best_acc:
        best_acc = acc
        best_feature = col
        best_mapping = mapping

print(f"Mejor atributo: {best_feature}")
print(f"Exactitud en entrenamiento: {best_acc:.4f}")
print("Regla generada (valor -> clase):")
for val, cls in best_mapping.items():
    print(f"  {best_feature} = {val}  ->  {cls}")

# ===== 3) Predecir en TEST usando la mejor regla =====
y_pred_test = X_test[best_feature].map(best_mapping).fillna(majority_class)

# ===== 4) Evaluar =====
print("\nEvaluación en Test:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_test):.4f}\n")
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred_test))

labels = sorted(pd.unique(pd.concat([y_train, y_test], ignore_index=True)))
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_test, labels=labels))
print("Labels:", labels)

```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer, make_column_selector as selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ===== 1) Configuración =====
TARGET = "ProductivityLoss"

df_train = df_limpio_entrenamiento_sin.copy()
df_test  = df_limpio_muestra_sin.copy()

X_train = df_train.drop(columns=[TARGET])
y_train = df_train[TARGET]

X_test  = df_test.drop(columns=[TARGET])
y_test  = df_test[TARGET]

# ===== 2) Preprocesamiento =====
# Cambia 'sparse_output=False' por 'sparse=False' si usas scikit-learn < 1.2
preprocess = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False),
         selector(dtype_include=["object", "category"])),
        ("num", "passthrough", selector(dtype_exclude=["object", "category"]))
    ],
    remainder="drop",
    verbose_feature_names_out=False
)

# ===== 3) Modelo Bayes =====
bayes = GaussianNB()

clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", bayes)
])

# ===== 4) Entrenar =====
clf.fit(X_train, y_train)

# ===== 5) Evaluar =====
y_pred = clf.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

labels = sorted(pd.unique(pd.concat([y_train, y_test], ignore_index=True)))
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred, labels=labels))
print("Labels:", labels)
```

# prism

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ===== 1) Configuración =====
TARGET = "ProductivityLoss"

train_df = df_limpio_entrenamiento_sin.copy()
test_df = df_limpio_muestra_sin.copy()

# ===== 2) Función PRISM =====
def prism_train(df, target):
    rules = []
    classes = df[target].unique()

    for clase in classes:
        df_clase = df.copy()
        # Mientras haya ejemplos de la clase
        while any(df_clase[target] == clase):
            condiciones = []
            subconjunto = df_clase.copy()

            # Construir una regla para esta clase
            while True:
                best_prec = -1
                best_cond = None

                for col in df_clase.columns:
                    if col == target:
                        continue
                    for val in df_clase[col].unique():
                        # Filtrar por la condición
                        df_temp = subconjunto[subconjunto[col] == val]
                        if len(df_temp) == 0:
                            continue
                        # Calcular precisión para esta clase
                        prec = sum(df_temp[target] == clase) / len(df_temp)
                        if prec > best_prec:
                            best_prec = prec
                            best_cond = (col, val)

                if best_cond is None:
                    break

                condiciones.append(best_cond)
                # Filtrar subconjunto
                col, val = best_cond
                subconjunto = subconjunto[subconjunto[col] == val]

                # Parar si todos son de la clase
                if all(subconjunto[target] == clase):
                    break

            # Guardar la regla
            rules.append((condiciones, clase))
            # Eliminar instancias cubiertas por la regla
            mask = pd.Series(True, index=df_clase.index)
            for col, val in condiciones:
                mask &= (df_clase[col] == val)
            df_clase = df_clase[~mask]

    return rules

# ===== 3) Entrenar PRISM =====
rules = prism_train(train_df, TARGET)

print("Reglas generadas:")
for conds, clase in rules:
    condiciones_str = " AND ".join([f"{col}={val}" for col, val in conds])
    print(f"SI {condiciones_str} ENTONCES {TARGET}={clase}")

# ===== 4) Función de predicción =====
def prism_predict(rules, df, default_class=None):
    preds = []
    for _, row in df.iterrows():
        pred = None
        for conds, clase in rules:
            if all(row[col] == val for col, val in conds):
                pred = clase
                break
        if pred is None:
            pred = default_class
        preds.append(pred)
    return preds

# Clase por defecto: la mayoritaria en el train
default_class = train_df[TARGET].value_counts().idxmax()

# ===== 5) Evaluar =====
y_pred = prism_predict(rules, test_df, default_class=default_class)
y_true = test_df[TARGET]

print(f"\nAccuracy: {accuracy_score(y_true, y_pred):.4f}\n")
print("Reporte de clasificación:")
print(classification_report(y_true, y_pred))

labels = sorted(pd.unique(pd.concat([y_true, pd.Series(y_pred)], ignore_index=True)))
print("Matriz de confusión:")
print(confusion_matrix(y_true, y_pred, labels=labels))
print("Labels:", labels)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth

# =============== CONFIG =================
MIN_SUPPORT = 0.05        # soporte mínimo en TRAIN (ajústalo)
MIN_CONF    = 0.60        # confianza mínima
NUM_BINS    = 5           # discretización para numéricas
RESTR_CONS  = "ProductivityLoss"  # pon None si no quieres restringir el consecuente

df_train = df_limpio_entrenamiento_sin.copy()
df_test  = df_limpio_muestra_sin.copy()

# =============== 1) PREPARACIÓN COMO TRANSACCIONES (one-hot) =================
def df_to_transaction_ohe(df: pd.DataFrame, num_bins=5) -> pd.DataFrame:
    """Convierte un DF 'wide' a matriz OHE de items 'col=valor' por fila (transacción).
       Discretiza numéricas en quantiles para convertirlas a categóricas."""
    tmp = df.copy()

    # 1. Discretizar numéricas a intervalos (quantiles)
    num_cols = tmp.select_dtypes(exclude=["object", "category", "bool"]).columns
    bin_edges = {}
    for c in num_cols:
        tmp[c], bins = pd.qcut(tmp[c], q=num_bins, duplicates="drop", retbins=True)
        bin_edges[c] = bins

    # 2. Convertir todas a string "col=valor"
    for c in tmp.columns:
        tmp[c] = tmp[c].astype(str).map(lambda v: f"{c}={v}")

    # 3. One-hot por transacción
    #    Stack -> get_dummies -> volver a agrupar por índice de fila
    ohe = pd.get_dummies(tmp.stack()).groupby(level=0).max()

    # Asegurar dtype booleano (mlxtend lo prefiere)
    ohe = ohe.astype(bool)

    return ohe

Xtr_ohe = df_to_transaction_ohe(df_train, num_bins=NUM_BINS)
Xte_ohe = df_to_transaction_ohe(df_test,  num_bins=NUM_BINS)

# =============== 2) MINADO DE ITEMSETS FRECUENTES (APRIORI/FP-GROWTH) =========
# Usa APRIORI (comenta y descomenta si prefieres FP-Growth)
freq = apriori(Xtr_ohe, min_support=MIN_SUPPORT, use_colnames=True)
# freq = fpgrowth(Xtr_ohe, min_support=MIN_SUPPORT, use_colnames=True)  # alternativa

# =============== 3) GENERACIÓN DE REGLAS EN TRAIN =============================
rules = association_rules(freq, metric="confidence", min_threshold=MIN_CONF)

# (Opcional) Restringe el consecuente a la clase, si te interesa reglas -> ProductivityLoss
if RESTR_CONS:
    rules = rules[rules["consequents"].apply(
        lambda s: any(item.startswith(f"{RESTR_CONS}=") for item in s)
    )]

# Orden inicial por (confidence * lift) para priorizar reglas fuertes
rules = rules.sort_values(by=["confidence", "lift", "support"], ascending=False).reset_index(drop=True)

print("Reglas (TRAIN) — columnas clave: antecedents, consequents, support, confidence, lift")
print(rules[["antecedents","consequents","support","confidence","lift"]].head(10))

# =============== 4) EVALUACIÓN DE REGLAS EN TEST ==============================
def support_on(df_ohe: pd.DataFrame, items: list[str]) -> float:
    """Soporte de un conjunto de items en una matriz OHE (proporción de filas con TODOS los items=1)."""
    if len(items) == 0:
        return 0.0
    # Si algún item no existe en test (columna ausente), el soporte es 0
    missing = [c for c in items if c not in df_ohe.columns]
    if missing:
        return 0.0
    return (df_ohe[items].all(axis=1)).mean()

def eval_rules_on_test(rules_df: pd.DataFrame, Xtest_ohe: pd.DataFrame) -> pd.DataFrame:
    rules_eval = rules_df.copy()
    s_test, c_test, l_test = [], [], []

    for _, r in rules_eval.iterrows():
        antecedents = list(r["antecedents"])
        consequents = list(r["consequents"])

        supp_ant_test  = support_on(Xtest_ohe, antecedents)
        supp_cons_test = support_on(Xtest_ohe, consequents)
        supp_both_test = support_on(Xtest_ohe, antecedents + consequents)

        conf_test = 0.0 if supp_ant_test == 0 else (supp_both_test / supp_ant_test)
        lift_test = 0.0 if supp_cons_test == 0 else (conf_test / supp_cons_test)

        s_test.append(supp_both_test)
        c_test.append(conf_test)
        l_test.append(lift_test)

    rules_eval["support_test"]    = s_test
    rules_eval["confidence_test"] = c_test
    rules_eval["lift_test"]       = l_test
    return rules_eval

rules_test = eval_rules_on_test(rules, Xte_ohe)

# Ordena por desempeño en TEST
rules_test = rules_test.sort_values(by=["confidence_test","lift_test","support_test"], ascending=False).reset_index(drop=True)

print("\nTOP reglas evaluadas en TEST — usa estas columnas para filtrar:")
cols_show = ["antecedents","consequents","support","confidence","lift","support_test","confidence_test","lift_test"]
print(rules_test[cols_show].head(10))
```
